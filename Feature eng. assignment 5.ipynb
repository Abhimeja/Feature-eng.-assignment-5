To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you would need a dataset that includes pairs of values for these two variables for each student. Let's denote the time spent studying as variable X and the exam scores as variable Y.

The Pearson correlation coefficient (often denoted as "r") can be calculated using the following formula:

Where:

Once you've calculated the correlation coefficient, its value will fall between -1 and 1:

If r is close to 1, it indicates a strong positive linear correlation, suggesting that as the time spent studying increases, the exam scores tend to increase as well.

If r is close to -1, it indicates a strong negative linear correlation, suggesting that as the time spent studying increases, the exam scores tend to decrease.

If r is close to 0, it indicates a weak or no linear correlation between the two variables.
